# Room Grid View (Window, Light, Fan)

```python
final = ""
for i in range(f_height):
    start = ' ------- '*f_width + "\n" + '|       |'*f_width+"\n"
    mid = ""
    for j in range(f_width):
        if j == 5:
            if i == 1 or i == 4 or i == 7:
                mid += "|  "+str(i)+","+str(j)+"  |"
            else:
                mid += "| "+str(i)+","+str(j)+",W |"
        elif (j == 1 or j == 2 or j == 3 or j == 4) and (i == 1 or i == 4 or i == 7 or i == 9):
            mid += "| "+str(i)+","+str(j)+",T |"
        elif j == 0 and (i == 1 or i == 8):
            mid += "| "+str(i)+","+str(j)+",G |"
        elif j == 0 and (i == 3 or i == 4 or i == 5 or i == 6):
            mid += "| "+str(i)+","+str(j)+",W |"
        elif (j == 1 or j == 2 or j == 3 or j == 4) and (i == 2 or i == 5 or i == 8):
            mid += "| "+str(i)+","+str(j)+",F |"
        else:
            mid += "|  "+str(i)+","+str(j)+"  |"
    end = '|       |'*f_width+"\n"+' ------- '*f_width+"\n"
    final += start + mid + "\n" + end
print(final)
```

```python
def getContour(s):
    print("Data\n")
    df = pd.read_excel(s)
    
    print(df)
    s_lst = []
    for i in range(f_height):
        r_l = []
        for j in range(f_width):
            get_lst = list(df[j])
            r_l.append(get_lst[i])
        s_lst.append(r_l)

    x = np.arange(0, 6, 1)
    y = np.arange(0, 10, 1)
    [X, Y] = np.meshgrid(x, y)


    fig, ax = plt.subplots(1, 1)
    plt.figure(figsize = [10,10])
    ax.contourf(X ,Y, s_lst)
    ax.set_title('2D Contour Plot - 1')
    ax.set_xlabel('Room Width')
    ax.set_ylabel('Room Height')
    plt.show()


    contours = plt.contour(X, Y, np.array(s_lst), 8, colors="Black")
    plt.clabel(contours, inline=True, fontsize=8)
    fig = plt.figure(figsize=[10, 10])
    ax = fig.gca()
    ax.set_title('2D Contour Plot - 2')
    ax.set_xlabel('Room Width')
    ax.set_ylabel('Room Height')
    plt.imshow(s_lst, extent=[0, 5, 0, 8], origin='lower',
               cmap='RdGy', alpha=0.5)
    plt.colorbar();


    plt.figure()
    fig = plt.figure(figsize= [10, 10])
    ax = fig.gca(projection="3d")
    ax.set_title('3D Contour Plot')
    ax.set_xlabel('Room Width')
    ax.set_ylabel('Room Height')
    ax.set_zlabel("Lux")
    surf = ax.plot_surface(X, Y, np.array(s_lst), cmap=cm.coolwarm)
    plt.show()
```
